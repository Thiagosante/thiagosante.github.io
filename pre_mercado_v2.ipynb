```python
# ===============================
# SISTEMA DE GR√ÅFICOS DE MERCADO - FORMATO BRASILEIRO
# ===============================

import yfinance as yf
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timezone, timedelta
import pytz

# ===============================
# CONFIGURA√á√ÉO DE FUSO HOR√ÅRIO E FORMATO
# ===============================
# Definir fuso hor√°rio do Brasil (considerando hor√°rio de Bras√≠lia)
brasil_tz = pytz.timezone('America/Sao_Paulo')
agora_brasil = datetime.now(brasil_tz)

print("=" * 80)
print(f"SISTEMA DE AN√ÅLISE DE MERCADO - {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print("=" * 80)

# ===============================
# CONFIGURA√á√ÉO DOS ATIVOS
# ===============================
ordem_ativos = [
    "Ambev S.A. (B3)",
    "B3 S.A. (B3)",
    "Vale S.A. (B3)",
    "Petrobras PN (B3)",
    "Petrobras ON (B3)",
    "Ita√∫ Unibanco (B3)",
    "Banco Bradesco (B3)",
    "E-Mini S&P 500 F",
    "E-Mini Dow Jones F",
    "E-Mini Nasdaq 100 F",
    "E-Mini Russell 2000 F",
    "√çndice S&P 500",
    "√çndice Dow Jones",
    "√çndice Nasdaq 100",
    "√çndice Russell 2000",
    "√çndice de Volatilidade(VIX)",
    "√çndice Bovespa",
    "Emerging Markets ETF",
    "Ouro (Gold) F",
    "Prata (Silver) F",
    "Cobre (Copper) F",
    "Petr√≥leo Bruto F",
    "Petr√≥leo Brent F",
    "G√°s Natural F",
    "FTSE 100(Londres)",
    "CAC 40 (Paris)",
    "DAX (Frankfurt)",
    "Nikkei 225 (T√≥quio)",
    "HANG SENG  Hong Kong)",
    "S&P/ASX 200 (Austr√°lia)",
    "Euro Stoxx 50 ",
    "D√≥lar/Real F",
    "EUA de 2 Anos F",
    "EUA de 5 Anos F",
    "EUA de 10 Anos F",
    "Iene Japon√™s versus D√≥lar Americano",
    "Bitcoin USD",
    "Ethereum USD",
    "MSCI Brazil EWZ ADR",
    "Dow Jones Brazil BR20 ADR",
    "Ita√∫ Unibanco ADR",
    "Petrobras ADR B",
    "Petrobras ADR A",
    "Banco Bradesco S.A. ADR",
    "Vale  ADR",
    "Nubank ADR"
]

tickers = {
"Ambev S.A. (B3)" : "ABEV3.SA",
"B3 S.A. (B3)" : "B3SA3.SA",
"Vale S.A. (B3)" : "VALE3.SA",
"Petrobras PN (B3)" : "PETR4.SA",
"Petrobras ON (B3)" : "PETR3.SA",
"Ita√∫ Unibanco (B3)" : "ITUB4.SA",
"Banco Bradesco (B3)" : "BBDC4.SA",
"E-Mini S&P 500 F" : "ES=F",
"E-Mini Dow Jones F" : "YM=F",
"E-Mini Nasdaq 100 F" : "NQ=F",
"E-Mini Russell 2000 F" : "RTY=F",
"√çndice S&P 500" : "^GSPC",
"√çndice Dow Jones" : "^DJI",
"√çndice Nasdaq 100" : "^NDX",
"√çndice Russell 2000" : "^RUT",
"√çndice de Volatilidade(VIX)" : "^VIX",
"√çndice Bovespa" : "^BVSP",
"Emerging Markets ETF" : "EEM",
"Ouro (Gold) F" : "GC=F",
"Prata (Silver) F" : "SI=F",
"Cobre (Copper) F" : "HG=F",
"Petr√≥leo Bruto F" : "CL=F",
"Petr√≥leo Brent F" : "BZ=F",
"G√°s Natural F" : "NG=F",
"FTSE 100(Londres)" : "^FTSE",
"CAC 40 (Paris)" : "^FCHI",
"DAX (Frankfurt)" : "^GDAXI",
"Nikkei 225 (T√≥quio)" : "^N225",
"HANG SENG  Hong Kong)" : "^HSI",
"S&P/ASX 200 (Austr√°lia)" : "^AXJO",
"Euro Stoxx 50 " : "^STOXX50E",
"D√≥lar/Real F" : "6L=F",
"EUA de 2 Anos F" : "ZT=F",
"EUA de 5 Anos F" : "ZF=F",
"EUA de 10 Anos F" : "ZN=F",
"Iene Japon√™s versus D√≥lar Americano" : "JPY=X",
"Bitcoin USD" : "BTC-USD",
"Ethereum USD" : "ETH-USD",
"MSCI Brazil EWZ ADR" : "EWZ",
"Dow Jones Brazil BR20 ADR" : "^BR20",
"Ita√∫ Unibanco ADR" : "ITUB",
"Petrobras ADR B" : "PBR",
"Petrobras ADR A" : "PBR-A",
"Banco Bradesco  ADR" : "BBD",
"Vale  ADR" : "VALE",
"Nubank ADR" : "NU",
"Banco Bradesco S.A. ADR": "BBD"
}

# ===============================
# FUN√á√ïES DE FORMATA√á√ÉO BRASILEIRA
# ===============================
def formatar_valor_br(valor):
    """Formata valores num√©ricos no padr√£o brasileiro: 1.234,56"""
    if pd.isna(valor) or valor is None:
        return "N/A"

    try:
        # Se for string, converter para float
        if isinstance(valor, str):
            # Remover caracteres n√£o num√©ricos
            valor = valor.replace('R$', '').replace('$', '').replace('%', '').strip()
            valor = valor.replace('.', '').replace(',', '.')

        valor_float = float(valor)

        # Formatar com 2 casas decimais
        if valor_float == 0:
            return "0,00"

        # Usar separador de milhar como ponto e decimal como v√≠rgula
        valor_abs = abs(valor_float)
        valor_formatado = f"{valor_abs:,.2f}"

        # Trocar pontos por v√≠rgulas e vice-versa
        valor_formatado = valor_formatado.replace(',', 'X').replace('.', ',').replace('X', '.')

        # Adicionar sinal negativo se necess√°rio
        if valor_float < 0:
            return f"-{valor_formatado}"
        else:
            return valor_formatado

    except (ValueError, TypeError) as e:
        return str(valor)

def formatar_percentual_br(valor):
    """Formata percentuais no padr√£o brasileiro: 1,23%"""
    if pd.isna(valor) or valor is None:
        return "N/A"

    try:
        valor_float = float(valor)
        valor_formatado = f"{valor_float:,.2f}".replace('.', ',')

        # Se j√° n√£o tiver % no final, adicionar
        if not valor_formatado.endswith('%'):
            valor_formatado = f"{valor_formatado}%"

        return valor_formatado
    except (ValueError, TypeError):
        return "N/A"

def converter_timestamp_br(timestamp):
    """Converte timestamp UNIX para hor√°rio brasileiro formatado"""
    if pd.isna(timestamp) or timestamp is None:
        return "N/A"

    try:
        # Converter timestamp para datetime UTC
        dt_utc = datetime.fromtimestamp(timestamp, tz=timezone.utc)

        # Converter para hor√°rio de Bras√≠lia
        dt_brasil = dt_utc.astimezone(brasil_tz)

        # Formatar no padr√£o brasileiro
        return dt_brasil.strftime('%d/%m/%Y %H:%M:%S')
    except (ValueError, TypeError, OSError):
        return "N/A"

def formatar_data_hora_br(dt):
    """Formata datetime no padr√£o brasileiro"""
    if pd.isna(dt) or dt is None:
        return "N/A"

    try:
        if isinstance(dt, (int, float)):
            return converter_timestamp_br(dt)
        elif isinstance(dt, datetime):
            if dt.tzinfo is None:
                dt = brasil_tz.localize(dt)
            else:
                dt = dt.astimezone(brasil_tz)
            return dt.strftime('%d/%m/%Y %H:%M:%S')
        else:
            return str(dt)
    except Exception:
        return "N/A"

# ===============================
# FUN√á√ÉO PARA BUSCAR DADOS
# ===============================
def buscar_dados_completo(tickers_dict):
    """Busca dados completos do mercado com tratamento robusto"""
    dados = []

    print("üîç Coletando dados do mercado...")

    for nome, simbolo in tickers_dict.items():
        try:
            ticker = yf.Ticker(simbolo)
            info = ticker.info

            # Obter hist√≥rico recente para dados alternativos
            try:
                hist = ticker.history(period="2d")
                if not hist.empty:
                    # Extrair dados do hist√≥rico
                    ultima_linha = hist.iloc[-1]
                    penultima_linha = hist.iloc[-2] if len(hist) > 1 else ultima_linha

                    # Calcular varia√ß√µes baseadas no hist√≥rico
                    fechamento_atual = ultima_linha['Close']
                    fechamento_anterior = penultima_linha['Close']
                    variacao_rs = fechamento_atual - fechamento_anterior
                    variacao_percent = (variacao_rs / fechamento_anterior) * 100
                else:
                    fechamento_atual = info.get('regularMarketPrice', 0)
                    fechamento_anterior = info.get('previousClose', 0)
                    variacao_rs = info.get('regularMarketChange', 0)
                    variacao_percent = info.get('regularMarketChangePercent', 0)

            except:
                fechamento_atual = info.get('regularMarketPrice', 0)
                fechamento_anterior = info.get('previousClose', 0)
                variacao_rs = info.get('regularMarketChange', 0)
                variacao_percent = info.get('regularMarketChangePercent', 0)

            # Coletar todos os dados poss√≠veis
            dados_ativo = {
                "Ativo": nome,
                "S√≠mbolo": simbolo,
                "√öltimo Pre√ßo": info.get('regularMarketPrice'),
                "Abertura": info.get('open'),
                "M√≠nima do Dia": info.get('dayLow'),
                "M√°xima do Dia": info.get('dayHigh'),
                "Fechamento Anterior": info.get('previousClose'),
                "Varia√ß√£o Di√°ria (R$)": info.get('regularMarketChange') or variacao_rs,
                "Varia√ß√£o Di√°ria (%)": info.get('regularMarketChangePercent') or variacao_percent,

                "Pr√©-mercado Pre√ßo": info.get('preMarketPrice'),
                "Pr√©-mercado (R$)": info.get('preMarketChange'),
                "Pr√©-mercado (%)": info.get('preMarketChangePercent'),
                "Hora Pr√©-mercado": info.get('preMarketTime'),

                "P√≥s-mercado Pre√ßo": info.get('postMarketPrice'),
                "P√≥s-mercado (R$)": info.get('postMarketChange'),
                "P√≥s-mercado (%)": info.get('postMarketChangePercent'),
                "Hora P√≥s-mercado": info.get('postMarketTime'),

                "Volume": info.get('volume'),
                "Volume M√©dio": info.get('averageVolume'),
                "Capitaliza√ß√£o": info.get('marketCap'),
                "Fonte": "yfinance",
                "Data Coleta": agora_brasil
            }

            dados.append(dados_ativo)

            # Feedback progressivo
            if dados_ativo["Varia√ß√£o Di√°ria (%)"] is not None:
                sinal = "üìà" if dados_ativo["Varia√ß√£o Di√°ria (%)"] >= 0 else "üìâ"
                print(f"  {sinal} {nome:25} | Var: {dados_ativo['Varia√ß√£o Di√°ria (%)']:7.2f}%")

        except Exception as e:
            print(f"  ‚ùå {nome:25} | Erro: {str(e)[:30]}...")
            # Adicionar dados m√≠nimos em caso de erro
            dados.append({
                "Ativo": nome,
                "S√≠mbolo": simbolo,
                "√öltimo Pre√ßo": 0,
                "Varia√ß√£o Di√°ria (%)": 0,
                "Pr√©-mercado (%)": 0,
                "P√≥s-mercado (%)": 0,
                "Fonte": "erro",
                "Data Coleta": agora_brasil
            })

    print(f"‚úÖ Coleta conclu√≠da: {len(dados)} ativos processados")
    return dados

# ===============================
# PROCESSAR DADOS E CRIAR DATAFRAME
# ===============================
# Buscar dados
dados_brutos = buscar_dados_completo(tickers)

# Criar DataFrame
df = pd.DataFrame(dados_brutos)

# Ordenar conforme a ordem especificada
df['Ordem'] = df['Ativo'].apply(
    lambda x: ordem_ativos.index(x) if x in ordem_ativos else 999
)
df = df.sort_values('Ordem').drop('Ordem', axis=1).reset_index(drop=True)

# ===============================
# CRIAR C√ìPIAS NUM√âRICAS PARA GR√ÅFICOS
# ===============================
# Criar c√≥pia com valores num√©ricos para os gr√°ficos
df_graficos = df.copy()

# Garantir que temos valores num√©ricos para os gr√°ficos
for col in ['Varia√ß√£o Di√°ria (%)', 'Pr√©-mercado (%)', 'P√≥s-mercado (%)']:
    if col in df_graficos.columns:
        df_graficos[f'{col}_num'] = pd.to_numeric(df_graficos[col], errors='coerce')
    else:
        df_graficos[f'{col}_num'] = 0

# ===============================
# FORMATA√á√ÉO BRASILEIRA PARA EXIBI√á√ÉO
# ===============================
print("\n" + "="*80)
print("FORMATANDO DADOS NO PADR√ÉO BRASILEIRO")
print("="*80)

# Formatar valores monet√°rios
colunas_monetarias = [
    '√öltimo Pre√ßo', 'Abertura', 'M√≠nima do Dia', 'M√°xima do Dia',
    'Fechamento Anterior', 'Varia√ß√£o Di√°ria (R$)',
    'Pr√©-mercado Pre√ßo', 'Pr√©-mercado (R$)',
    'P√≥s-mercado Pre√ßo', 'P√≥s-mercado (R$)'
]

for col in colunas_monetarias:
    if col in df.columns:
        df[col] = df[col].apply(formatar_valor_br)

# Formatar percentuais
colunas_percentuais = ['Varia√ß√£o Di√°ria (%)', 'Pr√©-mercado (%)', 'P√≥s-mercado (%)']
for col in colunas_percentuais:
    if col in df.columns:
        df[col] = df[col].apply(formatar_percentual_br)

# Formatar timestamps
colunas_tempo = ['Hora Pr√©-mercado', 'Hora P√≥s-mercado']
for col in colunas_tempo:
    if col in df.columns:
        df[col] = df[col].apply(converter_timestamp_br)

# Formatar volume e capitaliza√ß√£o
if 'Volume' in df.columns:
    df['Volume'] = df['Volume'].apply(lambda x: f"{int(x):,}".replace(',', '.') if pd.notna(x) else "N/A")

if 'Capitaliza√ß√£o' in df.columns:
    df['Capitaliza√ß√£o'] = df['Capitaliza√ß√£o'].apply(
        lambda x: f"R${int(x)/1e9:,.1f} bi".replace(',', 'X').replace('.', ',').replace('X', '.')
        if pd.notna(x) and x > 0 else "N/A"
    )

# Formatar data de coleta
df['Data Coleta'] = df['Data Coleta'].apply(formatar_data_hora_br)

# ===============================
# EXIBIR DATAFRAME FORMATADO
# ===============================
print("\nüìä RESUMO DO MERCADO (FORMATO BRASILEIRO):")
print(f"Data/Hora: {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print(f"Total de ativos: {len(df)}")
print("-" * 80)

# Mostrar DataFrame formatado
from IPython.display import display
display(df)

# ===============================
# GR√ÅFICO 1: VARIA√á√ÉO DI√ÅRIA DOS ATIVOS (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 1: VARIA√á√ÉO DI√ÅRIA DOS ATIVOS (%)")
print("="*80)

# Ordenar por varia√ß√£o di√°ria para o gr√°fico
df_var_diaria = df_graficos.sort_values('Varia√ß√£o Di√°ria (%)_num', ascending=False)

# Criar cores baseadas na dire√ß√£o
cores_var_diaria = []
for valor in df_var_diaria['Varia√ß√£o Di√°ria (%)_num']:
    if pd.isna(valor):
        cores_var_diaria.append('gray')
    elif valor > 0:
        cores_var_diaria.append('#00CC96')  # Verde
    elif valor < 0:
        cores_var_diaria.append('#EF553B')  # Vermelho
    else:
        cores_var_diaria.append('#636EFA')  # Azul

# Criar gr√°fico com Plotly
fig1 = go.Figure()

fig1.add_trace(go.Bar(
    x=df_var_diaria['Ativo'],
    y=df_var_diaria['Varia√ß√£o Di√°ria (%)_num'],
    marker_color=cores_var_diaria,
    text=[f"{x:+.2f}%" for x in df_var_diaria['Varia√ß√£o Di√°ria (%)_num']],
    textposition='outside', # Alterado para for√ßar a visibilidade do texto
    textfont=dict(color='black'), # Adicionado para melhor legibilidade
    hovertemplate=(
        "<b>%{x}</b><br>" +
        "Varia√ß√£o Di√°ria: %{y:+.2f}%<br>" +
        "√öltimo Pre√ßo: R$ %{customdata[0]}<br>" +
        "Abertura: R$ %{customdata[1]}<extra></extra>"

    ),
    customdata=np.column_stack((
        df_var_diaria['√öltimo Pre√ßo'].apply(formatar_valor_br),
        df_var_diaria['Abertura'].apply(formatar_valor_br)
    ))
))

fig1.update_layout(
    title={
        'text': f"VARIA√á√ÉO DI√ÅRIA DOS ATIVOS<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_title="Ativos",
    yaxis_title="Varia√ß√£o Di√°ria (%)",
    xaxis_tickangle=-45,
    height=600,
    showlegend=False,
    plot_bgcolor='white',
    yaxis=dict(
        gridcolor='lightgray',
        zeroline=True,
        zerolinecolor='black',
        zerolinewidth=2
    ),
  xaxis=dict(
        categoryorder='array',
        categoryarray=ordem_ativos, # Alterado para ordem_ativos conforme solicitado
        gridcolor='lightgray'
    )
)

# Adicionar linha em y=0
fig1.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

fig1.show()

# ===============================
# GR√ÅFICO 2: VARIA√á√ÉO NO PR√â-MERCADO (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 2: VARIA√á√ÉO NO PR√â-MERCADO (%)")
print("="*80)

# Filtrar ativos com dados de pr√©-mercado
df_pre = df_graficos.copy()
df_pre = df_pre[df_pre['Pr√©-mercado (%)_num'].notna()]

if len(df_pre) > 0:
    # Ordenar por varia√ß√£o no pr√©-mercado
    df_pre = df_pre.sort_values('Pr√©-mercado (%)_num', ascending=False)

    # Criar cores
    cores_pre = []
    for valor in df_pre['Pr√©-mercado (%)_num']:
        if pd.isna(valor):
            cores_pre.append('gray')
        elif valor > 0:
            cores_pre.append('#00CC96')  # Verde
        elif valor < 0:
            cores_pre.append('#EF553B')  # Vermelho
        else:
            cores_pre.append('#636EFA')  # Azul

    fig2 = go.Figure()

    fig2.add_trace(go.Bar(
        x=df_pre['Ativo'],
        y=df_pre['Pr√©-mercado (%)_num'],
        marker_color=cores_pre,
        text=[f"{x:+.2f}%" for x in df_pre['Pr√©-mercado (%)_num']],
        textposition='outside', # Alterado para for√ßar a visibilidade do texto
        textfont=dict(color='black'), # Adicionado para melhor legibilidade
        hovertemplate=(
            "<b>%{x}</b><br>" +
            "Pr√©-mercado: %{y:+.2f}%<br>" +
            "Pre√ßo Pr√©-mercado: R$ %{customdata[0]}<br>" +
            "Fechamento Anterior: R$ %{customdata[1]}<br>" +
            "Hora: %{customdata[2]}<extra></extra>"
        ),
        customdata=np.column_stack((
            df_pre['Pr√©-mercado Pre√ßo'].apply(formatar_valor_br),
            df_pre['Fechamento Anterior'].apply(formatar_valor_br),
            df_pre['Hora Pr√©-mercado'].apply(converter_timestamp_br)
        ))
    ))

    fig2.update_layout(
        title={
            'text': f"VARIA√á√ÉO NO PR√â-MERCADO<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
            'x': 0.5,
            'xanchor': 'center'
        },
        xaxis_title="Ativos",
        yaxis_title="Varia√ß√£o no Pr√©-mercado (%)",
        xaxis_tickangle=-45,
        height=500,
        showlegend=False,
        plot_bgcolor='white',
        yaxis=dict(
            gridcolor='lightgray',
            zeroline=True,
            zerolinecolor='black',
            zerolinewidth=2
        ),
        xaxis=dict(
            gridcolor='lightgray'
        )
    )

    fig2.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

    fig2.show()
else:
    print("‚ö†Ô∏è  Nenhum dado de pr√©-mercado dispon√≠vel para os ativos selecionados.")

# ===============================
# GR√ÅFICO 3: VARIA√á√ÉO NO P√ìS-MERCADO (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 3: VARIA√á√ÉO NO P√ìS-MERCADO (%)")
print("="*80)

# Filtrar ativos com dados de p√≥s-mercado
df_pos = df_graficos.copy()
df_pos = df_pos[df_pos['P√≥s-mercado (%)_num'].notna()]

if len(df_pos) > 0:
    # Ordenar por varia√ß√£o no p√≥s-mercado
    df_pos = df_pos.sort_values('P√≥s-mercado (%)_num', ascending=False)

    # Criar cores
    cores_pos = []
    for valor in df_pos['P√≥s-mercado (%)_num']:
        if pd.isna(valor):
            cores_pos.append('gray')
        elif valor > 0:
            cores_pos.append('#00CC96')  # Verde
        elif valor < 0:
            cores_pos.append('#EF553B')  # Vermelho
        else:
            cores_pos.append('#636EFA')  # Azul

    fig3 = go.Figure()

    fig3.add_trace(go.Bar(
        x=df_pos['Ativo'],
        y=df_pos['P√≥s-mercado (%)_num'],
        marker_color=cores_pos,
        text=[f"{x:+.2f}%" for x in df_pos['P√≥s-mercado (%)_num']],
        textposition='auto',
        hovertemplate=(
            "<b>%{x}</b><br>" +
            "P√≥s-mercado: %{y:+.2f}%<br>" +
            "Pre√ßo P√≥s-mercado: R$ %{customdata[0]}<br>" +
            "Fechamento Anterior: R$ %{customdata[1]}<br>" +
            "Hora: %{customdata[2]}<extra></extra>"
        ),
        customdata=np.column_stack((
            df_pos['P√≥s-mercado Pre√ßo'].apply(formatar_valor_br),
            df_pos['Fechamento Anterior'].apply(formatar_valor_br),
            df_pos['Hora P√≥s-mercado'].apply(converter_timestamp_br)
        ))
    ))

    fig3.update_layout(
        title={
            'text': f"VARIA√á√ÉO NO P√ìS-MERCADO<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
            'x': 0.5,
            'xanchor': 'center'
        },
        xaxis_title="Ativos",
        yaxis_title="Varia√ß√£o no P√≥s-mercado (%)",
        xaxis_tickangle=-45,
        height=500,
        showlegend=False,
        plot_bgcolor='white',
        yaxis=dict(
            gridcolor='lightgray',
            zeroline=True,
            zerolinecolor='black',
            zerolinewidth=2
        ),
        xaxis=dict(
            gridcolor='lightgray'
        )
    )

    fig3.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

    fig3.show()
else:
    print("‚ö†Ô∏è  Nenhum dado de p√≥s-mercado dispon√≠vel para os ativos selecionados.")

# ===============================
# ESTAT√çSTICAS RESUMIDAS
# ===============================
print("\n" + "="*80)
print("üìä ESTAT√çSTICAS RESUMIDAS")
print("="*80)

# Calcular estat√≠sticas b√°sicas
print(f"Data/Hora da an√°lise: {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print(f"Total de ativos analisados: {len(df_graficos)}")

# Varia√ß√£o Di√°ria
if 'Varia√ß√£o Di√°ria (%)_num' in df_graficos.columns:
    var_diaria = df_graficos['Varia√ß√£o Di√°ria (%)_num'].dropna()
    if len(var_diaria) > 0:
        positivos = sum(var_diaria > 0)
        negativos = sum(var_diaria < 0)
        neutros = sum(var_diaria == 0)

        print(f"\nüìà VARIA√á√ÉO DI√ÅRIA:")
        print(f"   Positivos (alta): {positivos} ({positivos/len(var_diaria)*100:.1f} %)")
        print(f"   Negativos (baixa): {negativos} ({negativos/len(var_diaria)*100:.1f} %)")
        print(f"   Neutros: {neutros} ({neutros/len(var_diaria)*100:.1f} %)")
        print(f"   M√©dia: {var_diaria.mean():+.2f}%")
        print(f"   M√°xima: {var_diaria.max():+.2f}%")
        print(f"   M√≠nima: {var_diaria.min():+.2f}%")

# Pr√©-mercado
if 'Pr√©-mercado (%)_num' in df_graficos.columns:
    var_pre = df_graficos['Pr√©-mercado (%)_num'].dropna()
    if len(var_pre) > 0:
        positivos_pre = sum(var_pre > 0)
        negativos_pre = sum(var_pre < 0)
        neutros_pre = sum(var_pre == 0)

        print(f"\nüåÖ PR√â-MERCADO:")
        print(f"   Ativos com dados: {len(var_pre)}")
        print(f"   Positivos: {positivos_pre} ({positivos_pre/len(var_pre)*100:.1f} %)")
        print(f"   Negativos: {negativos_pre} ({negativos_pre/len(var_pre)*100:.1f} %)")
        print(f"   Neutros: {neutros_pre} ({neutros_pre/len(var_pre)*100:.1f} %)")
        print(f"   M√©dia: {var_pre.mean():+.2f}%")
        print(f"   M√°xima: {var_pre.max():+.2f}%")
        print(f"   M√≠nima: {var_pre.min():+.2f}%")
    else:
        print(f"\nüåÖ PR√â-MERCADO:")
        print("   Nenhum dado de pr√©-mercado dispon√≠vel para os ativos selecionados.")

# P√≥s-mercado
if 'P√≥s-mercado (%)_num' in df_graficos.columns:
    var_pos = df_graficos['P√≥s-mercado (%)_num'].dropna()
    if len(var_pos) > 0:
        positivos_pos = sum(var_pos > 0)
        negativos_pos = sum(var_pos < 0)
        neutros_pos = sum(var_pos == 0)

        print(f"\nüåÉ P√ìS-MERCADO:")
        print(f"   Ativos com dados: {len(var_pos)}")
        print(f"   Positivos: {positivos_pos} ({positivos_pos/len(var_pos)*100:.1f} %)")
        print(f"   Negativos: {negativos_pos} ({negativos_pos/len(var_pos)*100:.1f} %)")
        print(f"   Neutros: {neutros_pos} ({neutros_pos/len(var_pos)*100:.1f} %)")
        print(f"   M√©dia: {var_pos.mean():+.2f}%")
        print(f"   M√°xima: {var_pos.max():+.2f}%")
        print(f"   M√≠nima: {var_pos.min():+.2f}%")
    else:
        print(f"\nüåÉ P√ìS-MERCADO:")
        print("   Nenhum dado de p√≥s-mercado dispon√≠vel para os ativos selecionados.")

print("\n" + "="*80)
print("‚úÖ AN√ÅLISE CONCLU√çDA!")
print("="*80)

# ===============================
# OP√á√ÉO PARA EXPORTAR OS DADOS
# ===============================
try:
    exportar = input("\nüíæ Deseja exportar os dados para CSV? (s/n): ").strip().lower()
    if exportar == 's':
        nome_arquivo = f"dados_mercado_br_{agora_brasil.strftime('%Y%m%d_%H%M%S')}.csv"

        # Exportar DataFrame formatado
        df.to_csv(nome_arquivo, index=False, encoding='utf-8-sig', sep=';')

        print(f"‚úÖ Dados exportados para: {nome_arquivo}")
        print(f"   Formato: CSV com separador ';' e encoding UTF-8-BOM")
        print(f"   Total de linhas: {len(df)}")
        print(f"   Total de colunas: {len(df.columns)}")
except:
    print("‚ö†Ô∏è  Modo n√£o interativo - pulando exporta√ß√£o")

print("\nüéØ RESUMO DOS GR√ÅFICOS GERADOS:")
print("   1. üìä Varia√ß√£o Di√°ria dos Ativos (%)")
print("   2. üåÖ Varia√ß√£o no Pr√©-mercado (%)")
print("   3. üåÉ Varia√ß√£o no P√≥s-mercado (%)")
print(f"\n‚è∞ An√°lise finalizada em: {datetime.now(brasil_tz).strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
```

```python
# ===============================
# SISTEMA DE GR√ÅFICOS DE MERCADO - FORMATO BRASILEIRO
# ===============================

import yfinance as yf
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timezone, timedelta
import pytz

# ===============================
# CONFIGURA√á√ÉO DE FUSO HOR√ÅRIO E FORMATO
# ===============================
# Definir fuso hor√°rio do Brasil (considerando hor√°rio de Bras√≠lia)
brasil_tz = pytz.timezone('America/Sao_Paulo')
agora_brasil = datetime.now(brasil_tz)

print("=" * 80)
print(f"SISTEMA DE AN√ÅLISE DE MERCADO - {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print("=" * 80)

# ===============================
# CONFIGURA√á√ÉO DOS ATIVOS
# ===============================
ordem_ativos = [
    "Ambev S.A. (B3)",
    "B3 S.A. (B3)",
    "Vale S.A. (B3)",
    "Petrobras PN (B3)",
    "Petrobras ON (B3)",
    "Ita√∫ Unibanco (B3)",
    "Banco Bradesco (B3)",
    "E-Mini S&P 500 F",
    "E-Mini Dow Jones F",
    "E-Mini Nasdaq 100 F",
    "E-Mini Russell 2000 F",
    "√çndice S&P 500",
    "√çndice Dow Jones",
    "√çndice Nasdaq 100",
    "√çndice Russell 2000",
    "√çndice de Volatilidade(VIX)",
    "√çndice Bovespa",
    "Emerging Markets ETF",
    "Ouro (Gold) F",
    "Prata (Silver) F",
    "Cobre (Copper) F",
    "Petr√≥leo Bruto F",
    "Petr√≥leo Brent F",
    "G√°s Natural F",
    "FTSE 100(Londres)",
    "CAC 40 (Paris)",
    "DAX (Frankfurt)",
    "Nikkei 225 (T√≥quio)",
    "HANG SENG  Hong Kong)",
    "S&P/ASX 200 (Austr√°lia)",
    "Euro Stoxx 50 ",
    "D√≥lar/Real F",
    "EUA de 2 Anos F",
    "EUA de 5 Anos F",
    "EUA de 10 Anos F",
    "Iene Japon√™s versus D√≥lar Americano",
    "Bitcoin USD",
    "Ethereum USD",
    "MSCI Brazil EWZ ADR",
    "Dow Jones Brazil BR20 ADR",
    "Ita√∫ Unibanco ADR",
    "Petrobras ADR B",
    "Petrobras ADR A",
    "Banco Bradesco S.A. ADR",
    "Vale  ADR",
    "Nubank ADR"
]

tickers = {
"Ambev S.A. (B3)" : "ABEV3.SA",
"B3 S.A. (B3)" : "B3SA3.SA",
"Vale S.A. (B3)" : "VALE3.SA",
"Petrobras PN (B3)" : "PETR4.SA",
"Petrobras ON (B3)" : "PETR3.SA",
"Ita√∫ Unibanco (B3)" : "ITUB4.SA",
"Banco Bradesco (B3)" : "BBDC4.SA",
"E-Mini S&P 500 F" : "ES=F",
"E-Mini Dow Jones F" : "YM=F",
"E-Mini Nasdaq 100 F" : "NQ=F",
"E-Mini Russell 2000 F" : "RTY=F",
"√çndice S&P 500" : "^GSPC",
"√çndice Dow Jones" : "^DJI",
"√çndice Nasdaq 100" : "^NDX",
"√çndice Russell 2000" : "^RUT",
"√çndice de Volatilidade(VIX)" : "^VIX",
"√çndice Bovespa" : "^BVSP",
"Emerging Markets ETF" : "EEM",
"Ouro (Gold) F" : "GC=F",
"Prata (Silver) F" : "SI=F",
"Cobre (Copper) F" : "HG=F",
"Petr√≥leo Bruto F" : "CL=F",
"Petr√≥leo Brent F" : "BZ=F",
"G√°s Natural F" : "NG=F",
"FTSE 100(Londres)" : "^FTSE",
"CAC 40 (Paris)" : "^FCHI",
"DAX (Frankfurt)" : "^GDAXI",
"Nikkei 225 (T√≥quio)" : "^N225",
"HANG SENG  Hong Kong)" : "^HSI",
"S&P/ASX 200 (Austr√°lia)" : "^AXJO",
"Euro Stoxx 50 " : "^STOXX50E",
"D√≥lar/Real F" : "6L=F",
"EUA de 2 Anos F" : "ZT=F",
"EUA de 5 Anos F" : "ZF=F",
"EUA de 10 Anos F" : "ZN=F",
"Iene Japon√™s versus D√≥lar Americano" : "JPY=X",
"Bitcoin USD" : "BTC-USD",
"Ethereum USD" : "ETH-USD",
"MSCI Brazil EWZ ADR" : "EWZ",
"Dow Jones Brazil BR20 ADR" : "^BR20",
"Ita√∫ Unibanco ADR" : "ITUB",
"Petrobras ADR B" : "PBR",
"Petrobras ADR A" : "PBR-A",
"Banco Bradesco  ADR" : "BBD",
"Vale  ADR" : "VALE",
"Nubank ADR" : "NU",
"Banco Bradesco S.A. ADR": "BBD"
}

# ===============================
# FUN√á√ïES DE FORMATA√á√ÉO BRASILEIRA
# ===============================
def formatar_valor_br(valor):
    """Formata valores num√©ricos no padr√£o brasileiro: 1.234,56"""
    if pd.isna(valor) or valor is None:
        return "N/A"

    try:
        # Se for string, converter para float
        if isinstance(valor, str):
            # Remover caracteres n√£o num√©ricos
            valor = valor.replace('R$', '').replace('$', '').replace('%', '').strip()
            valor = valor.replace('.', '').replace(',', '.')

        valor_float = float(valor)

        # Formatar com 2 casas decimais
        if valor_float == 0:
            return "0,00"

        # Usar separador de milhar como ponto e decimal como v√≠rgula
        valor_abs = abs(valor_float)
        valor_formatado = f"{valor_abs:,.2f}"

        # Trocar pontos por v√≠rgulas e vice-versa
        valor_formatado = valor_formatado.replace(',', 'X').replace('.', ',').replace('X', '.')

        # Adicionar sinal negativo se necess√°rio
        if valor_float < 0:
            return f"-{valor_formatado}"
        else:
            return valor_formatado

    except (ValueError, TypeError) as e:
        return str(valor)

def formatar_percentual_br(valor):
    """Formata percentuais no padr√£o brasileiro: 1,23%"""
    if pd.isna(valor) or valor is None:
        return "N/A"

    try:
        valor_float = float(valor)
        valor_formatado = f"{valor_float:,.2f}".replace('.', ',')

        # Se j√° n√£o tiver % no final, adicionar
        if not valor_formatado.endswith('%'):
            valor_formatado = f"{valor_formatado}%"

        return valor_formatado
    except (ValueError, TypeError):
        return "N/A"

def converter_timestamp_br(timestamp):
    """Converte timestamp UNIX para hor√°rio brasileiro formatado"""
    if pd.isna(timestamp) or timestamp is None:
        return "N/A"

    try:
        # Converter timestamp para datetime UTC
        dt_utc = datetime.fromtimestamp(timestamp, tz=timezone.utc)

        # Converter para hor√°rio de Bras√≠lia
        dt_brasil = dt_utc.astimezone(brasil_tz)

        # Formatar no padr√£o brasileiro
        return dt_brasil.strftime('%d/%m/%Y %H:%M:%S')
    except (ValueError, TypeError, OSError):
        return "N/A"

def formatar_data_hora_br(dt):
    """Formata datetime no padr√£o brasileiro"""
    if pd.isna(dt) or dt is None:
        return "N/A"

    try:
        if isinstance(dt, (int, float)):
            return converter_timestamp_br(dt)
        elif isinstance(dt, datetime):
            if dt.tzinfo is None:
                dt = brasil_tz.localize(dt)
            else:
                dt = dt.astimezone(brasil_tz)
            return dt.strftime('%d/%m/%Y %H:%M:%S')
        else:
            return str(dt)
    except Exception:
        return "N/A"

# ===============================
# FUN√á√ÉO PARA BUSCAR DADOS
# ===============================
def buscar_dados_completo(tickers_dict):
    """Busca dados completos do mercado com tratamento robusto"""
    dados = []

    print("üîç Coletando dados do mercado...")

    for nome, simbolo in tickers_dict.items():
        try:
            ticker = yf.Ticker(simbolo)
            info = ticker.info

            # Obter hist√≥rico recente para dados alternativos
            try:
                hist = ticker.history(period="2d")
                if not hist.empty:
                    # Extrair dados do hist√≥rico
                    ultima_linha = hist.iloc[-1]
                    penultima_linha = hist.iloc[-2] if len(hist) > 1 else ultima_linha

                    # Calcular varia√ß√µes baseadas no hist√≥rico
                    fechamento_atual = ultima_linha['Close']
                    fechamento_anterior = penultima_linha['Close']
                    variacao_rs = fechamento_atual - fechamento_anterior
                    variacao_percent = (variacao_rs / fechamento_anterior) * 100
                else:
                    fechamento_atual = info.get('regularMarketPrice', 0)
                    fechamento_anterior = info.get('previousClose', 0)
                    variacao_rs = info.get('regularMarketChange', 0)
                    variacao_percent = info.get('regularMarketChangePercent', 0)

            except:
                fechamento_atual = info.get('regularMarketPrice', 0)
                fechamento_anterior = info.get('previousClose', 0)
                variacao_rs = info.get('regularMarketChange', 0)
                variacao_percent = info.get('regularMarketChangePercent', 0)

            # Coletar todos os dados poss√≠veis
            dados_ativo = {
                "Ativo": nome,
                "S√≠mbolo": simbolo,
                "√öltimo Pre√ßo": info.get('regularMarketPrice'),
                "Abertura": info.get('open'),
                "M√≠nima do Dia": info.get('dayLow'),
                "M√°xima do Dia": info.get('dayHigh'),
                "Fechamento Anterior": info.get('previousClose'),
                "Varia√ß√£o Di√°ria (R$)": info.get('regularMarketChange') or variacao_rs,
                "Varia√ß√£o Di√°ria (%)": info.get('regularMarketChangePercent') or variacao_percent,

                "Pr√©-mercado Pre√ßo": info.get('preMarketPrice'),
                "Pr√©-mercado (R$)": info.get('preMarketChange'),
                "Pr√©-mercado (%)": info.get('preMarketChangePercent'),
                "Hora Pr√©-mercado": info.get('preMarketTime'),

                "P√≥s-mercado Pre√ßo": info.get('postMarketPrice'),
                "P√≥s-mercado (R$)": info.get('postMarketChange'),
                "P√≥s-mercado (%)": info.get('postMarketChangePercent'),
                "Hora P√≥s-mercado": info.get('postMarketTime'),

                "Volume": info.get('volume'),
                "Volume M√©dio": info.get('averageVolume'),
                "Capitaliza√ß√£o": info.get('marketCap'),
                "Fonte": "yfinance",
                "Data Coleta": agora_brasil
            }

            dados.append(dados_ativo)

            # Feedback progressivo
            if dados_ativo["Varia√ß√£o Di√°ria (%)"] is not None:
                sinal = "üìà" if dados_ativo["Varia√ß√£o Di√°ria (%)"] >= 0 else "üìâ"
                print(f"  {sinal} {nome:25} | Var: {dados_ativo['Varia√ß√£o Di√°ria (%)']:7.2f}%")

        except Exception as e:
            print(f"  ‚ùå {nome:25} | Erro: {str(e)[:30]}...")
            # Adicionar dados m√≠nimos em caso de erro
            dados.append({
                "Ativo": nome,
                "S√≠mbolo": simbolo,
                "√öltimo Pre√ßo": 0,
                "Varia√ß√£o Di√°ria (%)": 0,
                "Pr√©-mercado (%)": 0,
                "P√≥s-mercado (%)": 0,
                "Fonte": "erro",
                "Data Coleta": agora_brasil
            })

    print(f"‚úÖ Coleta conclu√≠da: {len(dados)} ativos processados")
    return dados

# ===============================
# PROCESSAR DADOS E CRIAR DATAFRAME
# ===============================
# Buscar dados
dados_brutos = buscar_dados_completo(tickers)

# Criar DataFrame
df = pd.DataFrame(dados_brutos)

# Ordenar conforme a ordem especificada
df['Ordem'] = df['Ativo'].apply(
    lambda x: ordem_ativos.index(x) if x in ordem_ativos else 999
)
df = df.sort_values('Ordem').drop('Ordem', axis=1).reset_index(drop=True)

# ===============================
# CRIAR C√ìPIAS NUM√âRICAS PARA GR√ÅFICOS
# ===============================
# Criar c√≥pia com valores num√©ricos para os gr√°ficos
df_graficos = df.copy()

# Garantir que temos valores num√©ricos para os gr√°ficos
for col in ['Varia√ß√£o Di√°ria (%)', 'Pr√©-mercado (%)', 'P√≥s-mercado (%)']:
    if col in df_graficos.columns:
        df_graficos[f'{col}_num'] = pd.to_numeric(df_graficos[col], errors='coerce')
    else:
        df_graficos[f'{col}_num'] = 0

# ===============================
# FORMATA√á√ÉO BRASILEIRA PARA EXIBI√á√ÉO
# ===============================
print("\n" + "="*80)
print("FORMATANDO DADOS NO PADR√ÉO BRASILEIRO")
print("="*80)

# Formatar valores monet√°rios
colunas_monetarias = [
    '√öltimo Pre√ßo', 'Abertura', 'M√≠nima do Dia', 'M√°xima do Dia',
    'Fechamento Anterior', 'Varia√ß√£o Di√°ria (R$)',
    'Pr√©-mercado Pre√ßo', 'Pr√©-mercado (R$)',
    'P√≥s-mercado Pre√ßo', 'P√≥s-mercado (R$)'
]

for col in colunas_monetarias:
    if col in df.columns:
        df[col] = df[col].apply(formatar_valor_br)

# Formatar percentuais
colunas_percentuais = ['Varia√ß√£o Di√°ria (%)', 'Pr√©-mercado (%)', 'P√≥s-mercado (%)']
for col in colunas_percentuais:
    if col in df.columns:
        df[col] = df[col].apply(formatar_percentual_br)

# Formatar timestamps
colunas_tempo = ['Hora Pr√©-mercado', 'Hora P√≥s-mercado']
for col in colunas_tempo:
    if col in df.columns:
        df[col] = df[col].apply(converter_timestamp_br)

# Formatar volume e capitaliza√ß√£o
if 'Volume' in df.columns:
    df['Volume'] = df['Volume'].apply(lambda x: f"{int(x):,}".replace(',', '.') if pd.notna(x) else "N/A")

if 'Capitaliza√ß√£o' in df.columns:
    df['Capitaliza√ß√£o'] = df['Capitaliza√ß√£o'].apply(
        lambda x: f"R${int(x)/1e9:,.1f} bi".replace(',', 'X').replace('.', ',').replace('X', '.')
        if pd.notna(x) and x > 0 else "N/A"
    )

# Formatar data de coleta
df['Data Coleta'] = df['Data Coleta'].apply(formatar_data_hora_br)

# ===============================
# EXIBIR DATAFRAME FORMATADO
# ===============================
print("\nüìä RESUMO DO MERCADO (FORMATO BRASILEIRO):")
print(f"Data/Hora: {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print(f"Total de ativos: {len(df)}")
print("-" * 80)

# Mostrar DataFrame formatado
from IPython.display import display
display(df)

# ===============================
# GR√ÅFICOS DE VARIA√á√ÉO DI√ÅRIA POR CATEGORIA (%)
# ===============================

print("\n" + "="*80)
print("GR√ÅFICOS DE VARIA√á√ÉO DI√ÅRIA POR CATEGORIA (%)")
print("="*80)

# Iterate through each category in ativos_por_categoria
for categoria, ativos_lista in ativos_por_categoria.items():
    print(f"\nGR√ÅFICO: VARIA√á√ÉO DI√ÅRIA PARA CATEGORIA: {categoria.upper()} (%)")
    print("-" * 80)

    # 2. Filter the df_graficos DataFrame for the current category
    df_categoria = df_graficos[df_graficos['Ativo'].isin(ativos_lista)].copy()

    # Check if there's any data for the category
    if df_categoria.empty:
        print(f"‚ö†Ô∏è Nenhum dado dispon√≠vel para a categoria '{categoria}'. Pulando gr√°fico.")
        continue

    # 3. Sort this filtered DataFrame by the 'Varia√ß√£o Di√°ria (%)_num' column
    df_categoria = df_categoria.sort_values('Varia√ß√£o Di√°ria (%)_num', ascending=False)

    # 4. Create a list of colors for the bars
    cores_categoria = []
    for valor in df_categoria['Varia√ß√£o Di√°ria (%)_num']:
        if pd.isna(valor):
            cores_categoria.append('gray')
        elif valor > 0:
            cores_categoria.append('#00CC96')  # Verde
        elif valor < 0:
            cores_categoria.append('#EF553B')  # Vermelho
        else:
            cores_categoria.append('#636EFA')  # Azul

    # 5. Initialize a Plotly go.Figure() object.
    fig_categoria = go.Figure()

    # 6. Add a bar trace using go.Bar()
    fig_categoria.add_trace(go.Bar(
        x=df_categoria['Ativo'],
        y=df_categoria['Varia√ß√£o Di√°ria (%)_num'],
        marker_color=cores_categoria,
        text=[f"{x:+.2f}%" for x in df_categoria['Varia√ß√£o Di√°ria (%)_num']],
        textposition='outside', # For√ßar a visibilidade do texto
        textfont=dict(color='black'), # Melhorar legibilidade
        hovertemplate=(
            "<b>%{x}</b><br>" +
            "Varia√ß√£o Di√°ria: %{y:+.2f}%<br>" +
            "√öltimo Pre√ßo: R$ %{customdata[0]}<br>" +
            "Abertura: R$ %{customdata[1]}<extra></extra>"
        ),
        customdata=np.column_stack((
            df_categoria['√öltimo Pre√ßo'].apply(formatar_valor_br),
            df_categoria['Abertura'].apply(formatar_valor_br)
        ))
    ))

    # 7. Update the figure's layout
    fig_categoria.update_layout(
        title={
            'text': f"VARIA√á√ÉO DI√ÅRIA {categoria.upper()}<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
            'x': 0.5,
            'xanchor': 'center'
        },
        xaxis_title="Ativos",
        yaxis_title="Varia√ß√£o Di√°ria (%)",
        xaxis_tickangle=-45,
        height=600,
        showlegend=False,
        plot_bgcolor='white',
        yaxis={
            'gridcolor': 'lightgray',
            'zeroline': True,
            'zerolinecolor': 'black',
            'zerolinewidth': 2
        },
        xaxis={
            'gridcolor': 'lightgray'
        }
    )

    # 8. Add a horizontal line at y=0
    fig_categoria.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

    # 9. Display the chart
    fig_categoria.show()

print("\n" + "="*80)
print("‚úÖ Gera√ß√£o de gr√°ficos por categoria conclu√≠da!")
print("="*80)

# ===============================
# GR√ÅFICO 2: VARIA√á√ÉO NO PR√â-MERCADO (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 2: VARIA√á√ÉO NO PR√â-MERCADO (%)")
print("="*80)

# Filtrar ativos com dados de pr√©-mercado
df_pre = df_graficos.copy()
df_pre = df_pre[df_pre['Pr√©-mercado (%)_num'].notna()]

if len(df_pre) > 0:
    # Ordenar por varia√ß√£o no pr√©-mercado
    df_pre = df_pre.sort_values('Pr√©-mercado (%)_num', ascending=False)

    # Criar cores
    cores_pre = []
    for valor in df_pre['Pr√©-mercado (%)_num']:
        if pd.isna(valor):
            cores_pre.append('gray')
        elif valor > 0:
            cores_pre.append('#00CC96')  # Verde
        elif valor < 0:
            cores_pre.append('#EF553B')  # Vermelho
        else:
            cores_pre.append('#636EFA')  # Azul

    fig2 = go.Figure()

    fig2.add_trace(go.Bar(
        x=df_pre['Ativo'],
        y=df_pre['Pr√©-mercado (%)_num'],
        marker_color=cores_pre,
        text=[f"{x:+.2f}%" for x in df_pre['Pr√©-mercado (%)_num']],
        textposition='outside', # Alterado para for√ßar a visibilidade do texto
        textfont=dict(color='black'), # Adicionado para melhor legibilidade
        hovertemplate=(
            "<b>%{x}</b><br>" +
            "Pr√©-mercado: %{y:+.2f}%<br>" +
            "Pre√ßo Pr√©-mercado: R$ %{customdata[0]}<br>" +
            "Fechamento Anterior: R$ %{customdata[1]}<br>" +
            "Hora: %{customdata[2]}<extra></extra>"
        ),
        customdata=np.column_stack((
            df_pre['Pr√©-mercado Pre√ßo'].apply(formatar_valor_br),
            df_pre['Fechamento Anterior'].apply(formatar_valor_br),
            df_pre['Hora Pr√©-mercado'].apply(converter_timestamp_br)
        ))
    ))

    fig2.update_layout(
        title={
            'text': f"VARIA√á√ÉO NO PR√â-MERCADO<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
            'x': 0.5,
            'xanchor': 'center'
        },
        xaxis_title="Ativos",
        yaxis_title="Varia√ß√£o no Pr√©-mercado (%)",
        xaxis_tickangle=-45,
        height=500,
        showlegend=False,
        plot_bgcolor='white',
        yaxis=dict(
            gridcolor='lightgray',
            zeroline=True,
            zerolinecolor='black',
            zerolinewidth=2
        ),
        xaxis=dict(
            gridcolor='lightgray'
        )
    )

    fig2.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

    fig2.show()
else:
    print("‚ö†Ô∏è  Nenhum dado de pr√©-mercado dispon√≠vel para os ativos selecionados.")

# ===============================
# GR√ÅFICO 3: VARIA√á√ÉO NO P√ìS-MERCADO (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 3: VARIA√á√ÉO NO P√ìS-MERCADO (%)")
print("="*80)

# Filtrar ativos com dados de p√≥s-mercado
df_pos = df_graficos.copy()
df_pos = df_pos[df_pos['P√≥s-mercado (%)_num'].notna()]

if len(df_pos) > 0:
    # Ordenar por varia√ß√£o no p√≥s-mercado
    df_pos = df_pos.sort_values('P√≥s-mercado (%)_num', ascending=False)

    # Criar cores
    cores_pos = []
    for valor in df_pos['P√≥s-mercado (%)_num']:
        if pd.isna(valor):
            cores_pos.append('gray')
        elif valor > 0:
            cores_pos.append('#00CC96')  # Verde
        elif valor < 0:
            cores_pos.append('#EF553B')  # Vermelho
        else:
            cores_pos.append('#636EFA')  # Azul

    fig3 = go.Figure()

    fig3.add_trace(go.Bar(
        x=df_pos['Ativo'],
        y=df_pos['P√≥s-mercado (%)_num'],
        marker_color=cores_pos,
        text=[f"{x:+.2f}%" for x in df_pos['P√≥s-mercado (%)_num']],
        textposition='auto',
        hovertemplate=(
            "<b>%{x}</b><br>" +
            "P√≥s-mercado: %{y:+.2f}%<br>" +
            "Pre√ßo P√≥s-mercado: R$ %{customdata[0]}<br>" +
            "Fechamento Anterior: R$ %{customdata[1]}<br>" +
            "Hora: %{customdata[2]}<extra></extra>"
        ),
        customdata=np.column_stack((
            df_pos['P√≥s-mercado Pre√ßo'].apply(formatar_valor_br),
            df_pos['Fechamento Anterior'].apply(formatar_valor_br),
            df_pos['Hora P√≥s-mercado'].apply(converter_timestamp_br)
        ))
    ))

    fig3.update_layout(
        title={
            'text': f"VARIA√á√ÉO NO P√ìS-MERCADO<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
            'x': 0.5,
            'xanchor': 'center'
        },
        xaxis_title="Ativos",
        yaxis_title="Varia√ß√£o no P√≥s-mercado (%)",
        xaxis_tickangle=-45,
        height=500,
        showlegend=False,
        plot_bgcolor='white',
        yaxis=dict(
            gridcolor='lightgray',
            zeroline=True,
            zerolinecolor='black',
            zerolinewidth=2
        ),
        xaxis=dict(
            gridcolor='lightgray'
        )
    )

    fig3.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

    fig3.show()
else:
    print("‚ö†Ô∏è  Nenhum dado de p√≥s-mercado dispon√≠vel para os ativos selecionados.")

# ===============================
# GR√ÅFICO 4: VARIA√á√ÉO DI√ÅRIA DOS ADRs BRASILEIROS (%)
# ===============================
print("\n" + "="*80)
print("GR√ÅFICO 4: VARIA√á√ÉO DI√ÅRIA DOS ADRs BRASILEIROS (%)")
print("="*80)

# Lista de ADRs brasileiros
adrs_brasileiros = [
    "MSCI Brazil EWZ ADR",
    "Dow Jones Brazil BR20 ADR",
    "Ita√∫ Unibanco ADR",
    "Petrobras ADR B",
    "Petrobras ADR A",
    "Banco Bradesco S.A. ADR",
    "Vale S.A. ADR",
    "Nubank ADR",

]

# Filtrar df_graficos para incluir apenas os ADRs brasileiros
df_adr_br = df_graficos[df_graficos['Ativo'].isin(adrs_brasileiros)].copy()

# Ordenar por varia√ß√£o di√°ria para o gr√°fico
df_adr_br = df_adr_br.sort_values('Varia√ß√£o Di√°ria (%)_num', ascending=False)

# Criar cores baseadas na dire√ß√£o
cores_adr_br = []
for valor in df_adr_br['Varia√ß√£o Di√°ria (%)_num']:
    if pd.isna(valor):
        cores_adr_br.append('gray')
    elif valor > 0:
        cores_adr_br.append('#00CC96')  # Verde
    elif valor < 0:
        cores_adr_br.append('#EF553B')  # Vermelho
    else:
        cores_adr_br.append('#636EFA')  # Azul

# Criar gr√°fico com Plotly
fig4 = go.Figure()

fig4.add_trace(go.Bar(
    x=df_adr_br['Ativo'],
    y=df_adr_br['Varia√ß√£o Di√°ria (%)_num'],
    marker_color=cores_adr_br,
    text=[f"{x:+.2f}%" for x in df_adr_br['Varia√ß√£o Di√°ria (%)_num']],
    textposition='auto',
    hovertemplate=(
        "<b>%{x}</b><br>" +
        "Varia√ß√£o Di√°ria: %{y:+.2f}%<br>" +
        "√öltimo Pre√ßo: R$ %{customdata[0]}<br>" +
        "Abertura: R$ %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack((
        df_adr_br['√öltimo Pre√ßo'].apply(formatar_valor_br),
        df_adr_br['Abertura'].apply(formatar_valor_br)
    ))
))

fig4.update_layout(
    title={
        'text': f"VARIA√á√ÉO DI√ÅRIA DOS ADRs BRASILEIROS<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_title="ADRs Brasileiros",
    yaxis_title="Varia√ß√£o Di√°ria (%)",
    xaxis_tickangle=-45,
    height=600,
    showlegend=False,
    plot_bgcolor='white',
    yaxis=dict(
        gridcolor='lightgray',
        zeroline=True,
        zerolinecolor='black',
        zerolinewidth=2
    ),
    xaxis=dict(
        categoryorder='array',
        categoryarray=ordem_ativos, # Alterado para ordem_ativos conforme solicitado
        gridcolor='lightgray'
    )
)

# Adicionar linha em y=0
fig4.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

fig4.show()

print("\n" + "="*80)
print("GR√ÅFICO 5: VARIA√á√ÉO DI√ÅRIA DOS FUTUROS DE COMMODITIES (%)")
print("="*80)

# 1. Define a Python list named `commodity_futures`
commodity_futures = [
    'Ouro (Gold) F',
    'Prata (Silver) F',
    'Cobre (Copper) F',
    'Petr√≥leo Bruto F',
    'Petr√≥leo Brent F',
    'G√°s Natural F'
    # 'Soja (Soybean) Futuros' - This asset is not present in the 'tickers' dictionary, so it cannot be included.
]

# 2. Filter the `df_graficos` DataFrame
df_commodity_futures = df_graficos[df_graficos['Ativo'].isin(commodity_futures)].copy()

# 3. Sort `df_commodity_futures`
df_commodity_futures = df_commodity_futures.sort_values('Varia√ß√£o Di√°ria (%)_num', ascending=False)

# 4. Create a list named `cores_commodity`
cores_commodity = []
for valor in df_commodity_futures['Varia√ß√£o Di√°ria (%)_num']:
    if pd.isna(valor):
        cores_commodity.append('gray')
    elif valor > 0:
        cores_commodity.append('#00CC96')  # Verde
    elif valor < 0:
        cores_commodity.append('#EF553B')  # Vermelho
    else:
        cores_commodity.append('#636EFA')  # Azul

# 5. Initialize a `go.Figure()` object.
fig5 = go.Figure()

# 6. Add a bar trace to the figure using `go.Bar()`
fig5.add_trace(go.Bar(
    x=df_commodity_futures['Ativo'],
    y=df_commodity_futures['Varia√ß√£o Di√°ria (%)_num'],
    marker_color=cores_commodity,
    text=[f"{x:+.2f}%" for x in df_commodity_futures['Varia√ß√£o Di√°ria (%)_num']],
    textposition='auto',
    hovertemplate=(
        "<b>%{x}</b><br>" +
        "Varia√ß√£o Di√°ria: %{y:+.2f}%<br>" +
        "√öltimo Pre√ßo: R$ %{customdata[0]}<br>" +
        "Abertura: R$ %{customdata[1]}<extra></extra>"
    ),
    customdata=np.column_stack((
        df_commodity_futures['√öltimo Pre√ßo'].apply(formatar_valor_br),
        df_commodity_futures['Abertura'].apply(formatar_valor_br)
    ))
))

# 7. Update the layout of the figure
fig5.update_layout(
    title={
        'text': f"VARIA√á√ÉO DI√ÅRIA DOS FUTUROS DE COMMODITIES<br><sup>{agora_brasil.strftime('%d/%m/%Y %H:%M')} (BRT)</sup>",
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_title="Commodities",
    yaxis_title="Varia√ß√£o Di√°ria (%)",
    xaxis_tickangle=-45,
    height=600,
    showlegend=False,
    plot_bgcolor='white',
    yaxis={
        'gridcolor': 'lightgray',
        'zeroline': True,
        'zerolinecolor': 'black',
        'zerolinewidth': 2
    },
    xaxis={
        'gridcolor': 'lightgray'
    }
)

# 8. Add a horizontal line at y=0
fig5.add_hline(y=0, line_dash="solid", line_color="black", opacity=0.7)

# 9. Display the chart
fig5.show()


# ===============================
# ESTAT√çSTICAS RESUMIDAS
# ===============================
print("\n" + "="*80)
print("üìä ESTAT√çSTICAS RESUMIDAS")
print("="*80)

# Calcular estat√≠sticas b√°sicas
print(f"Data/Hora da an√°lise: {agora_brasil.strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
print(f"Total de ativos analisados: {len(df_graficos)}")

# Varia√ß√£o Di√°ria
if 'Varia√ß√£o Di√°ria (%)_num' in df_graficos.columns:
    var_diaria = df_graficos['Varia√ß√£o Di√°ria (%)_num'].dropna()
    if len(var_diaria) > 0:
        positivos = sum(var_diaria > 0)
        negativos = sum(var_diaria < 0)
        neutros = sum(var_diaria == 0)

        print(f"\nüìà VARIA√á√ÉO DI√ÅRIA:")
        print(f"   Positivos (alta): {positivos} ({positivos/len(var_diaria)*100:.1f} %)")
        print(f"   Negativos (baixa): {negativos} ({negativos/len(var_diaria)*100:.1f} %)")
        print(f"   Neutros: {neutros} ({neutros/len(var_diaria)*100:.1f} %)")
        print(f"   M√©dia: {var_diaria.mean():+.2f}%")
        print(f"   M√°xima: {var_diaria.max():+.2f}%")
        print(f"   M√≠nima: {var_diaria.min():+.2f}%")

# Pr√©-mercado
if 'Pr√©-mercado (%)_num' in df_graficos.columns:
    var_pre = df_graficos['Pr√©-mercado (%)_num'].dropna()
    if len(var_pre) > 0:
        positivos_pre = sum(var_pre > 0)
        negativos_pre = sum(var_pre < 0)
        neutros_pre = sum(var_pre == 0)

        print(f"\nüåÖ PR√â-MERCADO:")
        print(f"   Ativos com dados: {len(var_pre)}")
        print(f"   Positivos: {positivos_pre} ({positivos_pre/len(var_pre)*100:.1f} %)")
        print(f"   Negativos: {negativos_pre} ({negativos_pre/len(var_pre)*100:.1f} %)")
        print(f"   Neutros: {neutros_pre} ({neutros_pre/len(var_pre)*100:.1f} %)")
        print(f"   M√©dia: {var_pre.mean():+.2f}%")
        print(f"   M√°xima: {var_pre.max():+.2f}%")
        print(f"   M√≠nima: {var_pre.min():+.2f}%")
    else:
        print(f"\nüåÖ PR√â-MERCADO:")
        print("   Nenhum dado de pr√©-mercado dispon√≠vel para os ativos selecionados.")

# P√≥s-mercado
if 'P√≥s-mercado (%)_num' in df_graficos.columns:
    var_pos = df_graficos['P√≥s-mercado (%)_num'].dropna()
    if len(var_pos) > 0:
        positivos_pos = sum(var_pos > 0)
        negativos_pos = sum(var_pos < 0)
        neutros_pos = sum(var_pos == 0)

        print(f"\nüåÉ P√ìS-MERCADO:")
        print(f"   Ativos com dados: {len(var_pos)}")
        print(f"   Positivos: {positivos_pos} ({positivos_pos/len(var_pos)*100:.1f} %)")
        print(f"   Negativos: {negativos_pos} ({negativos_pos/len(var_pos)*100:.1f} %)")
        print(f"   Neutros: {neutros_pos} ({neutros_pos/len(var_pos)*100:.1f} %)")
        print(f"   M√©dia: {var_pos.mean():+.2f}%")
        print(f"   M√°xima: {var_pos.max():+.2f}%")
        print(f"   M√≠nima: {var_pos.min():+.2f}%")
    else:
        print(f"\nüåÉ P√ìS-MERCADO:")
        print("   Nenhum dado de p√≥s-mercado dispon√≠vel para os ativos selecionados.")

print("\n" + "="*80)
print("‚úÖ AN√ÅLISE CONCLU√çDA!")
print("="*80)

# ===============================
# OP√á√ÉO PARA EXPORTAR OS DADOS
# ===============================
try:
    exportar = input("\nüíæ Deseja exportar os dados para CSV? (s/n): ").strip().lower()
    if exportar == 's':
        nome_arquivo = f"dados_mercado_br_{agora_brasil.strftime('%Y%m%d_%H%M%S')}.csv"

        # Exportar DataFrame formatado
        df.to_csv(nome_arquivo, index=False, encoding='utf-8-sig', sep=';')

        print(f"‚úÖ Dados exportados para: {nome_arquivo}")
        print(f"   Formato: CSV com separador ';' e encoding UTF-8-BOM")
        print(f"   Total de linhas: {len(df)}")
        print(f"   Total de colunas: {len(df.columns)}")
except:
    print("‚ö†Ô∏è  Modo n√£o interativo - pulando exporta√ß√£o")

print("\nüéØ RESUMO DOS GR√ÅFICOS GERADOS:")
print("   1. üìä Varia√ß√£o Di√°ria dos Ativos (%)")
print("   2. üåÖ Varia√ß√£o no Pr√©-mercado (%)")
print("   3. üåÉ Varia√ß√£o no P√≥s-mercado (%)")
print(f"\n‚è∞ An√°lise finalizada em: {datetime.now(brasil_tz).strftime('%d/%m/%Y %H:%M:%S')} (BRT)")
```
